# zip

zip() 函数用于将可迭代的对象作为参数，将对象中对应的元素打包成一个个元组，然后返回一个可迭代的zip对象.   
同时也可以有两个以上的参数;当传入参数的长度不同时，zip能自动以最短序列长度为准进行截取，获得元组。  
**如果要显示出来，用list( )函数**  
如果各个迭代器的元素个数不一致，则返回列表长度与最短的对象相同，利用 * 号操作符，可以将元组解压为列表。

- 可迭代对象、list显示：

```python
a = [1, 2, 3]
b = [4, 5, 6]
c = [4, 5, 6, 7, 8]

zipped = zip(a,b)
print(type(zipped))
for i in zipped:
    print(i)
list(zip(a,b))
```

```python
str1 = 'abc'
str2 = 'def123'
list(zip(str1, str2))
```

- 三个参数：

```python
l1,l2,l3 = (1,2,3),(4,5,6),(7,8,9)
list(zip(l1,l2,l3))
```

- 解包

```python
zipped = zip(l1, l2, l3)
zip_un = zip(*zipped)
list(zip_un)
```

# dict

- defaultdict  
使用defaultdict任何未定义的key都会默认返回一个默认的参数值, 而相同情况下dict()会返回KeyError.

```python
from collections import defaultdict
d2 = dict()
# d2['a']
d1 = defaultdict(int)
d3 = defaultdict(str)
v1 = d3['r']
v2 = d1['r']
print(v1)
print(v2)
```
